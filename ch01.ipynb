# Penguins
>knn은 거리를 기반으로 하기 때문에 문자열을 숫자로 바꿔줘야 한다

```python
import pandas as pd
```

```python
penguines = pd.read_csv('./data/penguins.csv')
penguines.info()
```

```python
penguines = penguines.dropna()
penguines.info()
```

```python
type(penguines)
```

```python
X = penguines[["bill_length_mm", 
              "bill_depth_mm",
              "flipper_length_mm",
              "body_mass_g"]] # dictonary이기 때문에 복합키를 사용할 수 없기 때문에
y = penguines['species']
```

```python
y.unique()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train) # 타겟은 숫자가 아니어도 잘 돈다
```

```python
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy # 예측정확도 -> 딥러닝 사용하면 +10%가 되겠구나라고 생각하자
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler() # 전처리 객체 만들었다. 표준분포 만들기
X_train_scaled = scalar.fit_transform(X_train) # 전처리해라. 학습모델이 아닌데 fit이 있음, fit은 무조건 trai넣는다는것!
X_test_scaled = scalar.fit_transform(X_test)

# 원래 이렇게 하는것
#
#
```

```python
knn.fit(X_train_scaled, y_train) # 학습 다시 해라
```

```python
y_pred = knn.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
accuracy # 전처리했더니 정확도가 올라갔다!
```

```python
import numpy as np

# 최적의 매개변수 찾기!!!!!
k_values = [1,3,5,7,9,11]
accuracies = []
for k in k_values:
    # 1. k에 따른 모델을 학습
    # 2. 예측
    # 3. 정확도 측정
    # 4. 저장
    # 5. 출력
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train_scaled, y_train)
    y_pred_temp = knn_temp.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred_temp)
    accuracies.append(acc)
    print(f"k={k}: 정확도: {acc:.4f}")

# 최적의 k를 찾으세요!
best_k = k_values[np.argmax(accuracies)]
best_k
```

```python
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
